```python

import pandas as pd
from glob import glob 


def get_csv(dir,nameofcsv):
    
    df =pd.read_csv("train.csv")
    img_path = glob(f"{dir}/*.png")
    img_path = sorted(img_path, key = lambda x: int(x.split("/")[1].split(".")[0].split("_")[1]))
    path_id= [x.split("/")[1] for x in img_path ]
    path_num =[x.split("_")[1].split(".")[0] for x in path_id]
    path_num = [int(x) for x in path_num]
    
    new_df = pd.DataFrame(columns = df.columns) 
    
    for num in path_num: 
        new_df = pd.concat([new_df,df[df.index==num]], axis= 0)
    new_df['id'] = path_id
    new_df['img_path'] =img_path
    
    new_df.to_csv(f"{nameofcsv}.csv")
    
    return None 
```

```python
import os 
import glob 
import pandas as pd 
import Augmentor

def get_augented_img(input_path, output_path):
    input_list = glob.glob(f"{input_path}+/*.jpg")+glob.glob(f"{input_path}+/*.png")
    input_list = sorted(input_list, key = lambda x : x.split("/")[-1].split("_")[1])
    
    p = Augmentor.Pipeline(input_path, output_path)
    p.flip_left_right(probability=0.9)
    p.rotate(probability=1, max_left_rotation=10, max_right_rotation=10)
    p.sample(len(input_list))
```
